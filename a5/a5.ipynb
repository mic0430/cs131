```python
# Upload the CSV file
from google.colab import files
uploaded = files.upload()

# Load the CSV file
filename = next(iter(uploaded))
data = pd.read_csv(filename, sep=',')

#  Convert
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
data['Goals'] = pd.to_numeric(data['Goals'], errors='coerce')

#Filter only forwards
def classify_position(position):
    if 'Forward' in position or 'Striker' in position:
        return 'Forward'
    else:
        return 'Other'

data['PositionCategory'] = data['Position'].apply(classify_position)
forwards_data = data[data['PositionCategory'] == 'Forward']

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(forwards_data['Age'], forwards_data['Goals'], alpha=0.7, c='blue')

#labels and title
plt.xlabel('Age')
plt.ylabel('Goals')
plt.title('Relationship Between Age and Goals for Forwards')
plt.grid(True)
plt.show()
```

```python
# Upload the CSV file
from google.colab import files
uploaded = files.upload()

# Load the CSV file
filename = next(iter(uploaded))
data = pd.read_csv(filename, sep=',')

#  Convert
data['MarketValue'] = pd.to_numeric(data['MarketValue'], errors='coerce')

# Convert MarketValue to millions of euros
data['MarketValue'] = data['MarketValue'] / 1e6

# SOrt them by
def classify_position(position):
    if 'Forward' in position or 'Striker' in position:
        return 'Forward'
    elif 'Midfield' in position:
        return 'Midfielder'
    elif 'Back' in position or 'Defender' in position:
        return 'Defender'
    elif 'Goalkeeper' in position:
        return 'Goalkeeper'
    else:
        return 'Other'


data['PositionCategory'] = data['Position'].apply(classify_position)


grouped_data = data.groupby('PositionCategory')['MarketValue'].agg(['min', 'max', 'mean']).reset_index()

#  Create a bar chart
plt.figure(figsize=(12, 8))

# Create bar positions
positions = range(len(grouped_data))

# Bar width
bar_width = 0.2

# Plotting
plt.bar(positions, grouped_data['min'], width=bar_width, label='Lowest Market Value')
plt.bar([p + bar_width for p in positions], grouped_data['mean'], width=bar_width, label='Mean Market Value')
plt.bar([p + bar_width*2 for p in positions], grouped_data['max'], width=bar_width, label='Highest Market Value')

# labels and title
plt.xlabel('Position Category')
plt.ylabel('Market Value (Millions of €)')
plt.title('Market Value Statistics by Position Category')
plt.xticks([p + bar_width for p in positions], grouped_data['PositionCategory'])
plt.legend()

# Display
plt.grid(True)
plt.show()
```
